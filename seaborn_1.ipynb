```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
#실선을 없애고 싶으면
sns.distplot(tips['total_bill'], kde=False)
```

```python
#막대그래프를 없애고 싶으면
sns.distplot(tips['total_bill'], hist=False)

#제목넣기
plt.title("Total Bill Density")
plt.xlabel('Total Bill')
plt.ylabel('Unit Probability')

plt.show()
```

```python
sns.distplot(tips['total_bill'], rug=True)
#rug 데이터가 있는지 없는지 보여주는 정도
```

```python
sns.countplot('day',data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
g=sns.jointplot(x='total_bill', y='tip', data=tips, alpha=0.5)
#alpha=투명도
g.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
g.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)

plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True)
```

```python
sns.barplot(x='time', y='total_bill', data=tips)
```

```python
#대각선으로 히스토그램이 그려짐
sns.pairplot(tips)
```

```python
#대각선이 히스토그램이 아니라 kde형태의 그래프로 그려짐
sns.pairplot(tips, diag_kind='kde')
```

```python
tips
```

```python
#그래프안에 내가 원하는대로 넣기
#대문자로 시작된것들은 클래스(객체)
#데이터들이 숫자인아이들만 나와. 그래서 3X3으로나옴
g = sns.PairGrid(tips)
g=g.map_upper(sns.regplot)  #대각선기준위쪽
g=g.map_lower(sns.kdeplot)   #대각선기준아래쪽
g=g.map_diag(sns.distplot, rug=True)   #대각선방향, 자기자신이랑비교, 큰의미는없는그래프

plt.show()
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris, hue='species')
#데이터들이 숫자인아이들만 나와. 그래서 4X4으로나옴
#hue='species' 색깔별로
```

```python
sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
```

```python
sns.pairplot(tips, hue='sex', diag_kind='kde')
```

```python
#그래프 크기 키우기 size옵션 추가
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', scatter_kws={'s':tips['size']*20})
#scatter_kws={}  키워드형태로
```

```python
#모양바꿔보기 'o','x','^','s','h'
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', scatter_kws={'s':tips['size']*20}, markers=['o','^'])
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)

plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set2')
#palette =box의 색상들 선택하는 옵션 (1,2,3)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='0.3')
```

```python
#두개를 한꺼번에 그려보기
sns.set_style('whitegrid')
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3')
sns.swarmplot(x='day', y='total_bill', data=tips, color='0.5')
```

```python
g = sns.FacetGrid(tips, col='time')
#time 은 lunch/dinner
g.map(sns.distplot, 'total_bill', rug=True)
```

```python
#요일별로 성별 분포
g = sns.FacetGrid(tips, col='day', hue='sex')
g = g.map(plt.scatter, 'total_bill', 'tip')
g = g.add_legend() #범례표현하기
```

```python
#담배피는사람 + 점심/저녁
g = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g = g.map(plt.scatter, 'total_bill', 'tip')
g = g.add_legend()
```
